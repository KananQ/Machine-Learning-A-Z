```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/diabetes.csv')
```

```python
df.head(3)
```

```python
df.plot()
```

```python
df.plot.line()
```

```python
#Scatter Plot
df.plot(kind='scatter',
        x='Age', y='Insulin')
```

```python
df.plot.scatter(x='Age', y='Insulin')
```

```python
#Histogram
#Syntax:plot(kind='hist) or plot.hist()
df.plot(kind='hist')
```

```python
df.plot.hist()
```

```python
#Area plot
#Syntax:plot(kind='area') or plot.area()
df.plot(kind='area')
```

```python
df.plot.area()
```

```python
#Bar Chart
#syntax: plot(kind='bar') or plot.bar()
#barh() for horizontal bar
df.plot.bar()
```

```python
df.plot.bar(stacked=True)
```

```python
df.plot.barh()
```

```python
#Box plot
#syntax plot(kind='box) or plot.box()
df.plot.box()
```

```python
#Hexagonal binning plot
#syntax plot.hexbin()
df.plot.hexbin(x='Glucose', y='BloodPressure',
               gridsize=18)
```

```python
#Histogram plot
#syntax df.plot.hist() or df.plot(kind='hist)
df.plot(kind='hist')
```

```python
df.plot.hist()
```

```python
#Density Estimate plot
#Syntax df.plot.kde() or plot.density()
df.plot.kde()
```

```python
#Pie plot
#Syntax plot.pie()
df.plot.pie(y='BloodPressure', autopct='%1.1f%%')
```
