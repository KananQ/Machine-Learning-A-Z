```python
#Ordinal definition
from pandas.api.types import CategoricalDtype
```

```python
df.cut.head()
```

```python
df.cut = df.cut.astype(CategoricalDtype(ordered = True))
```

```python
df.dtypes
```

```python
df.cut.head(1)
```

```python
cut_catagory = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
df.cut = df.cut.astype(CategoricalDtype(categories = cut_catagory, ordered = True))
```

```python
df.cut.head(1)
```

```python
#Barplot
df['cut'].value_counts().plot.barh();
```

```python
#Add set_title function:Add information to barplot
df['cut'].value_counts().plot.barh().set_title("About cut variables")
```

```python
#Write readable code strcture
(
  df['cut']
 .value_counts()
 .plot.barh()
 .set_title("About cut variables")
 )
```

```python
#Create plot with sns lib
sns.barplot(x = 'cut', y = df.cut.index, data=df);
```
